```python
reviews = pd.read_csv('/content/drive/MyDrive/Practice_datasets/winemag-data-130k-v2.csv')
reviews.head()
```

```python
reviews.points.describe()
```

```python
reviews.taster_name.unique()
```

```python
reviews.taster_name.value_counts()
```

```python
review_points_meaan = reviews.points.mean()
reviews.points.map(lambda p: p - review_points_meaan)
```

```python
def remean_points(row):
  row.points = row.points - review_points_meaan
  return row

reviews.apply(remean_points, axis='columns')
```

```python
reviews.points - review_points_meaan
```

```python
def star_review(row):
    if(row.country == 'Canada'):
        row['star_review'] = 3
    elif(row.points>=95):
        row['star_review'] = 3
    elif(row.points>=85):
        row['star_review'] = 2
    else:
        row['star_review'] = 1
    return row

star_ratings = reviews.apply(star_review, axis='columns').star_review
```
